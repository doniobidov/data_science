# Lab Exercise 3

## Ex 3.1

```python
# Fetch historical data for Nasdaq from 2010 to 2022
nasdaq_data = yf.download('^IXIC', start='2010-01-01', end='2022-12-31')

# Plotting the data
plt.figure(figsize=(6, 3))
plt.plot(nasdaq_data['Close'], label='Nasdaq Index')
plt.title('Nasdaq Stock Index (2010-2022)')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.legend()
plt.grid(True)
plt.show()
```

## Ex 3.2

```python
# Fetch historical data for Dow Jones from 2010 to 2022
dow_data = yf.download('^DJI', start='2010-01-01', end='2022-12-31')

# Plotting the data
plt.figure(figsize=(6, 3))

# Plot Nasdaq index
plt.plot(nasdaq_data['Close'], label='Nasdaq Index', color='blue')

# Plot Dow Jones index
plt.plot(dow_data['Close'], label='Dow Jones Index', color='red')

plt.title('Nasdaq and Dow Jones Stock Indices (2010-2022)')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.legend()
plt.grid(True)
plt.show()
```

## Ex 3.3

```python
# Compare the stock price of googl, meta, aapl, amzn, and tsla from 2010-2022
# Define the stock symbols
symbols = ['GOOGL', 'META', 'AAPL', 'AMZN', 'TSLA']

# Fetch historical data for each stock from 2010 to 2022
stock_data = yf.download(symbols, start='2010-01-01', end='2022-12-31')

# Plotting the data
plt.figure(figsize=(8, 4))

for symbol in symbols:
    plt.plot(stock_data['Close'][symbol], label=symbol, linewidth=1.2)

plt.title('Stock Prices Comparison (2010-2022)')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

# Lab Exercise 4

## Ex 4.1

```python
# URL of the World Bank data
world_bank_url = "https://api.worldbank.org/v2/en/indicator/NY.GDP.MKTP.CD?downloadformat=excel"

# Read data from Excel file
gdp_data = pd.read_excel(world_bank_url, sheet_name="Data", skiprows=3)

# Filter data for the United States and China
us_gdp = gdp_data[gdp_data["Country Name"] == "United States"].set_index("Country Name").transpose()
china_gdp = gdp_data[gdp_data["Country Name"] == "China"].set_index("Country Name").transpose()

# Filter data for the years 2010 to 2022
us_gdp_subset = us_gdp.loc["2010":"2022"]
china_gdp_subset = china_gdp.loc["2010":"2022"]

# Plotting the data
plt.figure(figsize=(8, 4))
plt.plot(us_gdp_subset.index, us_gdp_subset["United States"], label="US GDP", marker='o')
plt.plot(china_gdp_subset.index, china_gdp_subset["China"], label="China GDP", marker='o')

# Adding labels and title
plt.title('GDP Comparison: US vs China (2010-2022)')
plt.xlabel('Year')
plt.ylabel('GDP (Billion of US$)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Monthly expenses data
expenses = {
    'House Mortgage': 2000,
    'Car Lease': 500,
    'Water, Gas, Electricity': 400,
    'Gas': 200,
    'Food': 1100,
    'Leisure': 200,
    'Other Expenses': 600
}

# Calculate total expenses
total_expenses = sum(expenses.values())

# Plotting the pie chart
plt.figure(figsize=(6, 6))
plt.pie(expenses.values(), labels=expenses.keys(), autopct='%1.1f%%', startangle=140)
plt.title('Monthly Expenses Distribution')
plt.show()
```

## Ex 5.2

```python
# Sample data
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Plotting the bar chart
plt.figure(figsize=(8, 4))
bars = plt.bar(languages, popularity, color='skyblue')

# Adding text labels above each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom')

# Adding labels and title
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages')
plt.ylim(0, max(popularity) + 2)
plt.show()
```
