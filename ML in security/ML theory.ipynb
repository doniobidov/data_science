```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
square_feet = np.array([500, 690, 690, 700, 660, 690, 720, 705, 810, 703, 820, 510]).reshape(-1, 1)
price = np.array([800, 919, 928, 1092, 949, 938, 1102, 1150, 1230, 1110, 1200, 830])

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(square_feet, price, color='blue', label='Data Points')
plt.title('Price vs Square Feet')
plt.xlabel('Square Feet')
plt.ylabel('Price')
plt.grid(True)

# Linear Regression
model = LinearRegression()
model.fit(square_feet, price)

# Plotting the regression line
plt.plot(square_feet, model.predict(square_feet), color='red', label='Regression Line')

# Predicting price for 800 square feet
predicted_price = model.predict([[800]])
plt.scatter(800, predicted_price, color='green', label='Predicted Point')
plt.annotate(f'Predicted Price: ${predicted_price[0]:.2f}', (800, predicted_price), textcoords="offset points", xytext=(-20,25), ha='center')

plt.legend()
plt.show()
```
